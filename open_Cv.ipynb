```python
# showing img without matplotlip
img=cv2.imread('chill.jpg')
cv2.imshow('chill img',img)
grayImg=cv2.imread('chill.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('chill imgGray',grayImg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
face_clf=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
while True:
    ret, frame=cap.read()
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if ret==False:
        continue
    faces=face_clf.detectMultiScale(frame,1.05,5)
    for (x,y,w,h) in faces :
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),4)
        cv2.imshow("video frame", frame)      
# wait for user to press -q, then you will stop the loop
    key_pressed=cv2.waitKey(1)& 0xFF
    if key_pressed==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
